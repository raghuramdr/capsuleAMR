```python
#Gene presence absence data
gpa_df = pd.read_csv("/mnt/klebsiella_pneumoniae/results/gene_presence_absence_filt_length_frag.Rtab", sep="\t")
```

```python
gpa_df.head()
```

```python
gene_names = gpa_df["Gene"].tolist()
```

```python
gpa_df.drop(columns=["Gene"], inplace=True)
```

```python
gpa_df.head()
```

```python
gpa_df.T.index.name = "Genome ID"
```

```python
gpa_df=gpa_df.T.reset_index()
gpa_df.head()
```

```python
#Get the genome IDs per AMR. Fetch data for those genome IDs from the "data_col" dataframe
for amr in actual_genomes_df["Antibiotic"].unique():
    print(amr)
    label_mapping = []
    amr_df = actual_genomes_df[actual_genomes_df["Antibiotic"] == amr]
    print("The number of genomes after filtering for {} is {}".format(amr, amr_df.shape[0]))
    genomes_per_amr = amr_df["Genome ID"].unique()
    output_df = gpa_df.loc[gpa_df['Genome ID'].isin(genomes_per_amr)]
    print("The number of genomes in the GPA matrix for {} is {}".format(amr,output_df.shape[0]))
    labels_per_amr = amr_df.loc[amr_df["Genome ID"].isin(genomes_per_amr)][["Genome ID", "Resistant Phenotype"]]
    print(labels_per_amr.shape)
    full_output = output_df.merge(labels_per_amr, how="inner", on="Genome ID", suffixes=["_op", "_labels"])
    print(full_output.shape)
    amr = amr.replace(" ","_")
    print(amr)
    filename='staphylococcus_gpa_'+amr+'_label.csv'
    full_output.to_csv(os.path.join("/mnt/klebsiella_pneumoniae", filename) ,index=False)
             
    
                                                                               
```
